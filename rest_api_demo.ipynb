```python
import os
import json
import requests
from dotenv import load_dotenv

load_dotenv()

title = input('Введите фильм для поиска: ')
api_key = os.getenv('OMDB_API_KEY')

if not api_key:
    raise ValueError('Ключ OMDB_API_KEY не найден в .env')

response = requests.get(
    'https://www.omdbapi.com/', 
    params={
        't': title,
        'apikey': api_key
    },
    timeout=10
)
response.raise_for_status()
json_data = response.json()

if json_data.get('Response') != 'True':
    error_message = json_data.get('Error', 'Неизвестная ошибка OMDb API')
    raise ValueError(f'OMDb API вернул ошибку: {error_message}')

print(json.dumps(json_data, ensure_ascii=False, indent=2))

movie_title = json_data.get('Title')
year = json_data.get('Year')
genre = json_data.get('Genre')
director = json_data.get('Director')
imdb_rating = json_data.get('imdbRating')
actors = json_data.get('Actors')

print(
    f'Фильм "{movie_title}" ({year}) - жанр: {genre}. \n'
    f'Режиссёр: {director}. IMDb: {imdb_rating}. Актёры: {actors}.'
)
```
